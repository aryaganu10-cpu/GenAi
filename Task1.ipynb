## Preprocess Text Data

### Subtask:
Clean and preprocess the 'text' column by performing steps like lowercasing, removing punctuation, stop words, and tokenization to prepare it for topic modeling.


**Reasoning**:
To preprocess the text column, I will first get the English stop words list. Then, I will define a function to convert text to lowercase, remove punctuation, tokenize it, and filter out stop words and non-alphabetic tokens. Finally, I will apply this function to the 'text' column of `df_cleaned` to create a 'processed_text' column and display the head of the DataFrame.



```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    if not isinstance(text, str): # Handle non-string inputs
        return []
    # Lowercasing
    text = text.lower()
    # Remove punctuation
    text = ''.join([char for char in text if char not in string.punctuation])
    # Tokenization
    tokens = nltk.word_tokenize(text)
    # Remove stop words and non-alphabetic tokens
    processed_tokens = [word for word in tokens if word.isalpha() and word not in stop_words]
    return processed_tokens

df_cleaned['processed_text'] = df_cleaned['text'].apply(preprocess_text)
display(df_cleaned.head())
```

```python
import gensim.corpora as corpora
from gensim.models import LdaMulticore, CoherenceModel
import matplotlib.pyplot as plt

# Create Dictionary
id2word = corpora.Dictionary(df_cleaned['processed_text'])

# Create Corpus
texts = df_cleaned['processed_text']

# Term Document Frequency
corpus = [id2word.doc2bow(text) for text in texts]

print("Dictionary and Corpus created successfully.")
```

**Reasoning**:
Now that the dictionary and corpus are prepared, I will define a function `compute_coherence_values` to iterate through a range of topic numbers, train an LDA model for each, calculate its coherence score, and then call this function with an appropriate range to generate the coherence scores.



```python
def compute_coherence_values(dictionary, corpus, texts, limit, start=2, step=3):
    coherence_values = []
    model_list = []
    for num_topics in range(start, limit, step):
        model = LdaMulticore(corpus=corpus, id2word=dictionary, num_topics=num_topics, workers=2, passes=10, random_state=100)
        model_list.append(model)
        coherencemodel = CoherenceModel(model=model, texts=texts, dictionary=dictionary, coherence='c_v')
        coherence_values.append(coherencemodel.get_coherence())
    return model_list, coherence_values

# Compute coherence values for various number of topics
# You can adjust the limit and step based on your dataset size and desired granularity
model_list, coherence_values = compute_coherence_values(dictionary=id2word, corpus=corpus, texts=texts, start=2, limit=15, step=1)

print("Coherence values computed for different topic numbers.")
```

**Reasoning**:
To visually determine the optimal number of topics, I will plot the calculated coherence values against the corresponding number of topics. This will help identify the topic count that yields the highest coherence score.



```python
x = range(2, 15, 1)
plt.figure(figsize=(10, 6))
plt.plot(x, coherence_values)
plt.xlabel("Num Topics")
plt.ylabel("Coherence score")
plt.title("Coherence Score vs. Number of Topics")
plt.xticks(x)
plt.show()

# Find the optimal number of topics based on the highest coherence score
optimal_num_topics = x[coherence_values.index(max(coherence_values))]
print(f"The optimal number of topics is: {optimal_num_topics}")
```

## Perform Topic Modeling (LDA)

### Subtask:
Apply the Latent Dirichlet Allocation (LDA) model to the preprocessed 'text' data, using the optimal number of topics identified in the previous step, to extract underlying themes or topics.


**Reasoning**:
To apply the LDA model, I will initialize `gensim.models.LdaMulticore` with the `corpus`, `id2word`, and the `optimal_num_topics` found in the previous step. Then, I will train the model and print the top 10 keywords for each topic to understand the extracted themes.



```python
lda_model = LdaMulticore(corpus=corpus, id2word=id2word, num_topics=optimal_num_topics, random_state=100, passes=10, workers=2)

print("LDA Model Topics:")
for idx, topic in lda_model.print_topics(num_words=10):
    print(f"Topic: {idx}\nWords: {topic}\n")
```

## Analyze Topic Distribution by Message Type

### Subtask:
Examine the distribution of the identified topics across the different message types (labels '0' and '1'). This will involve associating each document's topic distribution with its corresponding label and calculating average topic probabilities per label.


**Reasoning**:
To analyze topic distribution, I will first get the topic distribution for each document using the trained LDA model, then convert these distributions into a pandas DataFrame with columns for each topic, ensuring that topics not present in a document are assigned a probability of 0.



```python
topic_distributions = []
for i, doc in enumerate(corpus):
    doc_topics = lda_model.get_document_topics(doc, minimum_probability=0) # Get all topics, even if probability is 0
    topic_dict = {f'Topic_{topic_id}': prob for topic_id, prob in doc_topics}
    topic_distributions.append(topic_dict)

# Convert list of dictionaries to DataFrame
topic_dist_df = pd.DataFrame(topic_distributions).fillna(0)

print("Topic distributions extracted and converted to DataFrame.")
```

**Reasoning**:
To associate each document's topic distribution with its corresponding label, I will add the 'label' column from `df_cleaned` to the `topic_dist_df` DataFrame. Then, to determine the average topic probabilities for each message type, I will group this combined DataFrame by the 'label' column and calculate the mean for each topic probability.



```python
topic_dist_df['label'] = df_cleaned['label'].values
average_topic_distribution = topic_dist_df.groupby('label').mean()
print("Average Topic Distribution by Message Type:")
print(average_topic_distribution)
```
