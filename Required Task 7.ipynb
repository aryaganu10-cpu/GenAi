## Comparison of Prompting Techniques and Conclusion

### Strengths and Weaknesses of Each Prompting Technique:

**1. Simple prompt (RTF Framework)**
*   **Strengths:** Simple, fast, concise, good for clear tasks.
*   **Weaknesses:** Lacks transparency, prone to inaccuracy/hallucination, limited depth.

**2. Chain of Thought (CoT)**
*   **Strengths:** Improves reasoning, better accuracy, explainable step-by-step process.
*   **Weaknesses:** Longer prompts, increased output length, errors can propagate.

**3. Tree of Thought (ToT)**
*   **Strengths:** Explores multiple angles, structured analysis, robust conclusions.
*   **Weaknesses:** Complex prompting, very verbose output, computationally intensive, overkill for simple tasks.

**4. Atom of Thought Approach**
*   **Strengths:** Granular, verifiable claims, high transparency, reduced bias, structured evaluation.
*   **Weaknesses:** Requires careful question formulation, can be repetitive, may miss broader context.

### Conclusion:

While a direct prompt, like the, offers efficiency for basic queries, its lack of transparency and depth can be limiting for complex analysis. Techniques such as Chain of Thought, Tree of Thought, and Atom of Thought provide increasingly structured and verifiable methods. These advanced approaches enhance analytical rigor and reduce bias by breaking down problems systematically, making them crucial when depth, explainability, and verifiability are paramount. The chosen technique should always match the task's complexity and required level of scrutiny.

#RTF

Prompt

Role: Act as an analyst for a commercial bank. Task: Analyse the attached webpage. Is Coinbase's strategy to reduce reliance on volatile trading fees actually working as of Q3 2024? Format: Return a 200 word summary https://www.fool.com/earnings/call-transcripts/2024/10/30/coinbase-global-coin-q3-2024-earnings-call-transcr/

Output:
200-word summary

As of Q3 2024, Coinbase’s strategy to reduce reliance on volatile trading fees appears to be partially working, though the business still remains meaningfully exposed to market-driven transaction revenue. Management stated it has made “a big effort to diversify…away from transaction fee revenue,” shifting toward subscription and services income, and is now “on pace to surpass 2 billion” from this segment in 2024, up from $1.4 billion in 2023. This demonstrates tangible progress in building more recurring, less cyclical revenue streams such as staking, custody, stablecoin monetization, and Coinbase One subscriptions.

In Q3 specifically, subscription and services revenue reached 556 million, close to transaction revenue of 573 million, indicating a more balanced mix than historically dominated by trading fees. Continued growth in USDC revenue, custody activity, and paid subscriptions further supports diversification.

However, the results also show the transition is incomplete. Transaction revenue still declined 27% quarter-on-quarter due to lower volatility, illustrating persistent sensitivity to market conditions. Even subscription revenue fell 7% sequentially because it remains indirectly linked to crypto prices and interest rates.

Overall, Coinbase’s diversification strategy is working structurally—improving earnings stability and creating a more balanced revenue base—but trading activity continues to be a major driver of quarterly performance.

##Tree of Thought

Prompt:

Approach: Tree of Thought (explore multiple angles, then converge) STEP 1 — Extract evidence (root facts) - Pull 10–15 short, verbatim-or-near-verbatim facts from the transcript. - Prioritize: segment revenues (transaction vs subscription/services), QoQ/YoY changes, management targets, drivers (staking, USDC, custody, subscriptions), and any commentary about volatility/volume sensitivity. - For each fact, tag it: [Revenue Mix], [Growth/Decline], [Drivers], [Risk/Dependency], [Guidance/Outlook].

STEP 2 — Build 4 analysis branches (angles) Branch A: Revenue mix / concentration risk - Assess: share of transaction revenue vs subscription & services; how close are they; direction of change. - Decision rule: diversification “working” if non-transaction is stable/growing and meaningfully offsets transaction swings. Branch B: Earnings stability / cyclicality - Assess: evidence of reduced volatility in results (e.g., profitability metrics mentioned, sequential sensitivity to volatility). - Decision rule: “working” if results look less tied to volatility/volume than before. Branch C: Unit economics & product drivers - Assess: which non-trading products are scaling (USDC, staking, custody, subscriptions) and whether growth is structural or market-linked. - Decision rule: “working” if drivers are recurring and not primarily price/volatility dependent. Branch D: Forward durability & risks - Assess: what could reverse progress (interest rates, regulatory, competitive fees, crypto price dependence). - Decision rule: “working” only if risks are acknowledged and the model still looks resilient under weaker volumes.

STEP 3 — Score each branch For each branch, assign a score from -2 to +2: -2 = clearly not working -1 = weak progress 0 = mixed +1 = working with caveats +2 = clearly working Provide 2–3 bullet reasons per score, each tied to an extracted fact.

STEP 4 — Synthesize (convergence) - Combine branch scores into an overall verdict: “Working”, “Partially working”, or “Not working”. - State 3 most important supporting facts and 2 biggest caveats. ##Atom of Thought Approach

Role: You are a commercial bank equity/credit analyst.

Question:
Is Coinbase’s strategy to reduce reliance on volatile trading fees actually working as of Q3 2024?

Source (use ONLY this webpage; do not use any other information):
https://www.fool.com/earnings/call-transcripts/2024/10/30/coinbase-global-coin-q3-2024-earnings-call-transcr/

Method: Atom-of-Thought (evaluate micro-claims one at a time)

Step 1 — Define the hypothesis:
“H1: Coinbase is successfully reducing reliance on trading fees.”

Step 2 — Create atomic evaluation questions (answer each independently using direct evidence):
A1: What proportion of revenue comes from transaction vs subscription & services in Q3?
A2: Has subscription & services revenue grown structurally vs prior periods?
A3: Did transaction revenue fall when volatility declined?
A4: Are non-transaction revenues recurring (staking, USDC, custody, subscriptions)?
A5: Are these non-transaction revenues still sensitive to crypto prices or interest rates?
A6: Has profitability improved or stabilized despite weaker trading conditions?

For each atom:
• Provide the exact supporting fact or number from the transcript.
• Label it as Supports H1, Weakly Supports H1, or Contradicts H1.

Step 3 — Aggregate atom results:
• Count how many atoms support vs contradict the hypothesis.
• Determine final classification: Working / Partially Working / Not Working.

Step 4 — Final Output:
Produce:
1) One-sentence verdict.
2) A 200-word evidence-based analyst summary citing only transcript data.
